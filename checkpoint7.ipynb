```python
import numpy as np
import pandas as pd
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes'],
'labels' : ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']}
df = pd.DataFrame(exam_data)
df=df.set_index('labels')
print(df)
#q1
df.head(3)
#q2
df = df.dropna()
print(df)
#q3
df.loc[:,["name","score"]]
print(df)
#q4
dict1={"name" : ["Suresh"], "score": [15.5], "attempts": [1], "qualify": ["yes"]}
df1=pd.DataFrame(dict1)
df=pd.concat([df,df1])
print(df)
#q5
df.drop("attempts", axis=1 , inplace=True)
print(df)
#q6
success=[]
for i in df["score"]:
    if float(i) > 10:
        success.append("1")
    else:
        success.append("0")
df["Success"]= success
print(df)
#q7
df.to_csv("df.csv")
df2=pd.read_csv("df.csv")
df2.head() #juste pour visualiser quelques lignes apr√®s l'exportationc
```
